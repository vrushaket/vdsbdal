```python
print(data['Car'].value_counts().idxmax())
data['Car'].replace(np.nan, 0, inplace=True)
```

```python
avg_buildingArea = data["BuildingArea"].astype("float").mean()
data["BuildingArea"].replace(np.nan, avg_buildingArea, inplace = True)

avg_YearBuilt = data["YearBuilt"].astype("float").median()
data["YearBuilt"].replace(np.nan, avg_YearBuilt, inplace = True)

data["CouncilArea"].replace(np.nan, "Bayside", inplace = True)
```

## Turn categorical variables into quantitative variables in Python

```python
categorical_columns = data.select_dtypes(include=['object'])
categorical_columns
```

```python
categorical_dummies = pd.get_dummies(categorical_columns)
categorical_dummies
```

```python
data = data.drop(list(categorical_columns.columns), axis=1)
data = pd.concat([data, categorical_dummies], axis=1)
```
