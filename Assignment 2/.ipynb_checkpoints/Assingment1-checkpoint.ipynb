### Data Preprocessing

Check for missing values in the data using pandas isnull()

```python
data.isnull().sum()
```

```python
columns = []
for cols in data:
    columns.append(cols)
data.dropna(subset=columns, how='any')

data.isnull().sum()
```

### Data Formatting and Data Normalization

Summarize the types of variables by checking the data types (i.e., character, numeric, integer, factor, and logical) of the variables in the data set.

```python
print(data.dtypes)
```

If variables are not in the correct data type, apply proper type conversions.

```python
# for cols in data:
#     data[cols] = data[cols].astype('float', errors='ignore')
# print(data.dtypes)  
```

```python
#data = data.apply(pd.to_numeric)
for cols in data:
    for value in data[cols]:
        if type(value) != data[cols].dtype:
            data[cols] = data[cols].astype(data[cols].dtype, errors='ignore')
            
```
